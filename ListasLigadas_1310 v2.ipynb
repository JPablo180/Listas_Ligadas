```python
class Nodo:
  def __init__( self, value, siguiente = None ):
    self.data = value
    self.next = siguiente
```

```python
head = Nodo( 10 )
print(head.data)
```

```python
head = Nodo( 10 )
head.next = Nodo( 20 )
print(f'Primero: { head.data }')
print(f'Segundo: {head.next.data} ')


# Lo mismo pero ahorrando cÃ³digo
head = Nodo( 10, Nodo(20))
print(f'Primero: {head.data }')
print(f'Segundo: {head.next.data}')

# Transversal
curr_node = head
while curr_node != None:
  print(f" | {curr_node.data} | --> ", end="")
  curr_node = curr_node.next
print("")
```

```python
class LinkedList:
  def __init__ (self):
    self.head = None
  
  def is_empty (self):
    return self.head == None
  
  def get_tail(self):
      if self.is_empty():
        return None
      else:  
        curr_node = self.head
        while curr_node.next != None:
          curr_node = curr_node.next
        return curr_node
  def append (self , value):
    if self.head == None:
        self.head = Nodo (value)
    else:
      curr_node = self.head
      while curr_node.next != None:
          curr_node = curr_node.next
      curr_node.next = Nodo(value)
  def tranversal(self):
    curr_node = self.head
    while curr_node != None:
      print(f'| {curr_node.data} | -->', end="")   
      curr_node = curr_node.next
    print("")

  def remove (self , value):
    if self.head == None:
      print('Lista vacia, no se puede remover nada')
    else:
      curr_node = self.head
      aux = None
      while curr_node.data != value : # revisar cuando el valor no exista
        aux = curr_node
        curr_node = curr_node.next
      
      if curr_node != None:
        aux.next = curr_node.next
        curr_node = None

  def preppend( self , value):
    if self.head == None:
      print('Esta vacio, no se puede hacer el prepend')
    else:
      aux = self.head
      self.head = Nodo ( value, aux)
  
  def pop(self):
    curr_node = self.head
    while curr_node.next.next != None:
      curr_node = curr_node.next
    print(f'El ulimo nodo es {curr_node.next.data} y este se va a eliminar ')
    curr_node.next = None 

  def pop_first(self):
    print(f'El primer nodo es {self.head.data} y este se va a eliminar')
    self.head = self.head.next


```

```python
lista = LinkedList()
if lista.is_empty():
  print('La lista esta vacia')
else:
  print('La lista no estÃ¡ vacia')

lista.append(10)
lista.append(20)
lista.append(30)
lista.append(40)
lista.append(50)

nodo = lista.get_tail()
if nodo != None:
  print(nodo.data)  
else:
  print('La lista esta vacia')

lista.tranversal() 
lista.remove(30)
lista.tranversal()
lista.preppend(5)
print('Despues del prepend')
lista.tranversal()
print('Usando Pop')
lista.pop()
lista.tranversal()
print('Usando el Pop_First')
lista.pop_first()
lista.tranversal()
```

```python
class NodoDoble:
  def __init__(self, value , siguiente = None, previo = None):
    self.data = value
    self.next = siguiente
    self.prev = previo
```

# Ejercicio con la clase nodo doble
- Crear un Nodo 10
  - Imprimir el valor del nodo
- Crear dos nodos
  - Imprimir el valor del primero
  - Imprimir el valor del segundo
- Lo mismo pero con 4 nodos 10, 20 ,30 ,40
  - Imprimir el valor del primero
  - Imprimir el valor del ultimo
  - Transversal
  - Reverse transversal
  - Insertar el nodo al final
  

```python
#ejercicio 1
head = NodoDoble(10)
tail = head
print(head.data)
print(tail.data)

#Ejercicio 2
print("Ejercicio 2: ")
head = NodoDoble(10, NodoDoble(20) )
head.next.prev = head
tail = head.next
print(f'El primer nodo es: {head.data}')
print(f'El primer nodo es: {tail.prev.data}')
print(f'El segundo nodo es: {head.next.data}')
print(f'El segundo nodo es: { tail.data}')


#Ejercicio 3
print('Ejericio 3: ')
head = NodoDoble(10, NodoDoble(20, NodoDoble(30, NodoDoble(40))))

tail = head.next.next.next
head.next.prev = head
head.next.next.prev = head.next
tail.prev = head.next.next


print(f'El primer nodo es: {head.data}')
print(f'El primer nodo es {tail.prev.prev.prev.data}')
print(f'El ultimo nodo es {tail.data}')
print(f'El primer nodo es {head.next.next.next.data}')

print("Agregar el 50 al final")
tail.next = NodoDoble(50 , None, tail)
tail = tail.next

# Transversal
curr_node = head
while curr_node != None:
  print(f" | {curr_node.data} | --> ", end="")
  curr_node = curr_node.next
print("")

#reverse Transversal
curr_node = tail
while curr_node != None:
  print(f" | {curr_node.data} | --> ", end="")
  curr_node = curr_node.prev
print("")

```

```python
class DoubleLinkedList:
  def __init__ (self) :
    self.head = None
    self.tail = None
    self.size = 0

  def get_size (self):
    return self.size  
  
  def is_empty ( self):
    return self.size == 0

  def append ( self, value):
    if self.is_empty():
      self.head = NodoDoble( value )
      self.tail = self.head
    else:
      self.tail.next = NodoDoble( value, None , self.tail)
      self.tail = self.tail.next
    self.size +=1

  def find_from_head(self, value):
      curr_node = self.head 
      contador = 0
      while curr_node.data != value:
          contador += 1
          curr_node = curr_node.next
      if curr_node.data == value:
          print(curr_node.data)


  def find_from_tail(self, value):
      curr_node = self.tail
      contador = 0
      while curr_node.data != value:
            contador += 1
            curr_node = curr_node.prev
        
      if curr_node.data == value:
        print(curr_node.data)       

  def remove_from_head(self, value):
      curr_node = self.head 
      while curr_node.data != value: 
            curr_node = curr_node.next
      if curr_node.data == value:
            curr_node.prev.next = curr_node.next
            curr_node.next.prev = curr_node.prev    


  def remove_from_tail( self , value):
      curr_node = self.tail
      while curr_node.data != value:
            curr_node = curr_node.prev
      if curr_node.data == value:
            curr_node.next.prev = curr_node.prev 
            curr_node.prev.next = curr_node.next      

  def transversal (self):
    curr_node = self.head
    while curr_node != None :
      print(f'| {curr_node.data} | --> ', end="")
      curr_node = curr_node.next
    print("")

  def reverse_transversal(self):
      curr_node = self.tail
      while curr_node != None:
          print(f"| {curr_node.data} | -->", end="")
          curr_node = curr_node.prev
      print("")    
```

```python
ldl = DoubleLinkedList()
ldl.transversal()
print("Con append")
ldl.append(10)
ldl.append(20)
ldl.append(30)
ldl.append(40)
ldl.transversal()
print(f'Esta vacia? {ldl.is_empty()} ')
print("\n")


print("Con find from head")
ldl.find_from_head(20)
ldl.transversal()
print("\n")

print("Con find from tail")
ldl.find_from_tail(10)
ldl.transversal()
print("\n")

print("Con remove_from_head")
ldl.remove_from_head(30)
ldl.transversal()
print("\n")


print("Con remove_from_tail")
ldl.remove_from_tail(20)
ldl.transversal()
print("\n")



print("Agregando de nuevo nodos")
ldl.append(50)
ldl.append(60)
ldl.append(70)
ldl.transversal()
print("\n")

print("Con get_size")
print(f"La Lista Doblemente Ligada tiene {ldl.get_size()} elementos")
print("\n")


print("Con transversal")
ldl.transversal()
print("\n")

print("Con reverse_transversal")
ldl.reverse_transversal()
```
