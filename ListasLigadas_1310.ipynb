## Todo esto antes de ir a la estructura ADT lista ligada

```python
class Nodo:
  def __init__( self, value, siguiente = None ):
    self.data = value
    self.next = siguiente
```

```python
head = Nodo( 10 )
print(head.data)
```

```python
head = Nodo( 10 )
head.next = Nodo( 20 )
print(f'Primero: { head.data }')
print(f'Segundo: {head.next.data} ')


# Lo mismo pero ahorrando cÃ³digo
head = Nodo( 10, Nodo(20))
print(f'Primero: {head.data }')
print(f'Segundo: {head.next.data}')

# Transversal
curr_node = head
while curr_node != None:
  print(f" | {curr_node.data} | --> ", end="")
  curr_node = curr_node.next
print("")
```

```python
head = Nodo (10, Nodo (20, Nodo (30 , Nodo(40))))
# Nodo (valor,siguiente)
# Transversal
curr_node = head
while curr_node != None:
  print(f"| {curr_node.data} | --> ", end= "")
  curr_node = curr_node.next
print("")


#imprimir el primero
print(head.data)
#imprimir el ultimo
curr_node = head
while curr_node.next != None:
    curr_node = curr_node.next
print(curr_node.data)

#insertar 25 despues del nodo 20 y antes del 30
curr_node = head
while curr_node.data != 20:
  curr_node = curr_node.next
print(curr_node.data)

# Crear el nodo 25
curr_node.next = Nodo(25, curr_node.next)

curr_node = head
while curr_node != None:
  print(f"| {curr_node.data} | --> ", end= "")
  curr_node = curr_node.next
print("")
```

```python
class LinkedList:
  def __init__ (self):
    self.head = None
  
  def is_empty (self):
    return self.head == None
  
  def get_tail(self):
      if self.is_empty():
        return None
      else:  
        curr_node = self.head
        while curr_node.next != None:
          curr_node = curr_node.next
        return curr_node
  def append (self , value):
    if self.head == None:
        self.head = Nodo (value)
    else:
      curr_node = self.head
      while curr_node.next != None:
          curr_node = curr_node.next
      curr_node.next = Nodo(value)
  def tranversal(self):
    curr_node = self.head
    while curr_node != None:
      print(f'| {curr_node.data} | -->', end="")   
      curr_node = curr_node.next
    print("")

  def remove (self , value):
    if self.head == None:
      print('Lista vacia, no se puede remover nada')
    else:
      curr_node = self.head
      aux = None
      while curr_node.data != value : # revisar cuando el valor no exista
        aux = curr_node
        curr_node = curr_node.next
      
      if curr_node != None:
        aux.next = curr_node.next
        curr_node = None

  def preppend( self , value):
    if self.head == None:
      print('Esta vacio, no se puede hacer el prepend')
    else:
      aux = self.head
      self.head = Nodo ( value, aux)
```

```python
lista = LinkedList()
if lista.is_empty():
  print('La lista esta vacia')
else:
  print('La lista no estÃ¡ vacia')

lista.append(10)
lista.append(20)
lista.append(30)
lista.append(40)
lista.append(50)

nodo = lista.get_tail()
if nodo != None:
  print(nodo.data)  
else:
  print('La lista esta vacia')

lista.tranversal() 
lista.remove(30)
lista.tranversal()
lista.preppend(5)
print('Despues del prepend')
lista.tranversal()
```
